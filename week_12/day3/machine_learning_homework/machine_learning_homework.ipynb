```python
cars_clean = cars.drop(columns = ['name'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
cars_clean['year'] = cars_clean['year'].astype('category')
```

```python
cars_clean = pd.get_dummies(cars_clean, drop_first = True)
cars_clean.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#cars_clean_ss = ss.fit_transform(cars_clean)
#cars_clean_ss = pd.DataFrame(cars_clean, columns = cars.columns)
#cars_clean_ss
```

```python
y = cars_clean['selling_price']
X = cars_clean.drop(columns = 'selling_price')
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
import statsmodels.api as sm
```

```python
x_i = sm.add_constant(X)
x_i.head()
```

```python
sm_model = sm.OLS(y, X.astype(float)).fit()
print(sm_model.summary())
```

```python
pd.DataFrame({
    'variable': X.columns.values,
    'coeffs': model.coef_
})
```

```python
residuals = sm_model.fittedvalues - y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = sm_model.fittedvalues, y = residuals, alpha = 0.2)
plt.xlabel('fitted value')
plt.ylabel('residual value')
```

```python
import scipy as sp
sp.stats.probplot(residuals, dist = 'norm', plot = plt)
```

Residuals are not normally distributed especially at the beginning and residuals do not show independency between each other. Looks like something else affects the price.
